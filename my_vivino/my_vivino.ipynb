```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
```

***
## le vin Rouge
***

```python
def Importation_de_vin_rouge():
    V_rouge = pd.read_csv("qualité_du_vin_rouge.csv" , sep=';')
    return V_rouge
dataset = Importation_de_vin_rouge()
```

***
## Résumer le jeu de données
***

```python
def Résumer_le_jeu_de_données(dataset):
    print("Les 10 premières lignes :")
    print(dataset.head(10))
    print("Dimension du jeu de données :")
    print(dataset.shape)
    print("Résumé statistique :")
    print(dataset.describe())
    print("Distribution des qualité :")
    print(dataset.groupby("quality").size())
#Résumer_le_jeu_de_données(dataset)
```

***
## nettoyer_dataset : 
Fonction qui supprime les valeurs aberrantes ou indésirables (comme 0, 0.0, …) 
et masque la colonne « qualité » pour préparer le jeu de données à l’analyse.
***

```python
def nettoyer_dataset(dataset):
    dataset_clean = dataset[(dataset != 0).all(axis=1)]
    dataset_clean = dataset_clean.drop(columns=["quality"])
    return dataset_clean
dataset_clean = nettoyer_dataset(dataset)
```

***
## la visualisation de données
***

```python
def Le_graphique(dataset):
    plt.hist(dataset["quality"], bins=6, color="#260A46FF")
    plt.xlabel("Évaluations de la qualité")
    plt.ylabel("Le recensement")
    plt.title("Graphique de répartition de la qualité")
    plt.show()
#Le_graphique(dataset)
```

***
## scaling :
permet de mettre toutes les variables sur la même 
échelle afin de comparer  les données  de manière  équitable
***

```python
def appliquer_scaling(dataset):
    scaler = MinMaxScaler()
    dataset_scaler = scaler.fit_transform(dataset)
    dataset_scaler = pd.DataFrame(dataset_scaler , index=dataset.index , columns=dataset.columns)
    return dataset_scaler
dataset_scaling = appliquer_scaling(dataset_clean)
```

***
## Calcul de similarité
Cette fonction calcule la matrice de similarité entre les vins en utilisant la cosine similarity afin d’identifier les vins les plus proches en termes de caractéristiques.
***

```python
def calculer_similarite_cosinus(dataset):
    matrice_de_similarité = cosine_similarity(dataset)
    similarite_dataset = pd.DataFrame(matrice_de_similarité , index=dataset.index , columns=dataset.index)
    return similarite_dataset
similarite_dataset = calculer_similarite_cosinus(dataset_clean)
```

***
## Recommander des vins
***

```python
def recommander_des_vins(vins_ID, Top_N = 5):
    similarite = similarite_dataset[vins_ID]
    recommendations = similarite.sort_values(ascending=False)
    recommendations = recommendations.drop(vins_ID)
    return recommendations.head(Top_N)
recommander_des_vins(10)
```

***
## Résultats de meilleure qualité
***

```python
def meilleure_qualité(vins_ID, Top_N = 5):
    meilleure = dataset[dataset["quality"] >= 7]
    meilleure = meilleure.drop(columns=["quality"])
    meilleure_scaling = appliquer_scaling(meilleure)
    meilleure_similarite = calculer_similarite_cosinus(meilleure_scaling)
    meilleure_ID = meilleure_similarite[vins_ID]
    meilleure_ID = meilleure_ID.sort_values(ascending=False)
    meilleure_ID = meilleure_ID.drop(vins_ID)
    return meilleure_ID.head(Top_N)
meilleure_qualité(7)
# Résultats de meilleure 
# dataset.loc[[8]]
```
